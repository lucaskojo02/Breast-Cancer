```python
#loading the important packages
from sklearn.datasets import load_breast_cancer
import matplotlib.pyplot as plt
from sklearn.inspection import DecisionBoundaryDisplay
from sklearn.svm import SVC
```

```python
#loading the breast cancer dataset
cancer = load_breast_cancer()
X = cancer.data[:, :2]  # we only take the first two features.
y = cancer.target
```

```python
#creating the SVC model
svm = SVC(kernel='rbf', C=1.0, gamma=0.5)
```

```python
#fitting the model
svm.fit(X, y)
```

```python
#Plotting the decision boundary
DecisionBoundaryDisplay.from_estimator(svm, X, response_method="predict", cmap=plt.cm.coolwarm, alpha=0.8,xlabel=cancer.feature_names[0], ylabel=cancer.feature_names[1])
```
